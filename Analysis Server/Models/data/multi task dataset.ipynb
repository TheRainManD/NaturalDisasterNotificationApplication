# Datasets from CrisisNLP

## Earthquake

```python
crisisnlp_earthquake_df = pd.read_csv("CrisisNLP/earthquakes_crisisnlp.tsv", sep='\t')
```

```python
crisisnlp_earthquake_df = crisisnlp_earthquake_df[crisisnlp_earthquake_df['lang'] == 'en']
```

```python
crisisnlp_earthquake_df = crisisnlp_earthquake_df[['text', 'label']]
crisisnlp_earthquake_df['label'].replace({'non-eyewitness': '1', "don't know": '0', 'indirect-eyewitness': '1', 
                                          'direct-eyewitness': '1'}, inplace = True)
crisisnlp_earthquake_df.drop(4, axis = 0, inplace = True)
```

```python
crisisnlp_earthquake_df['Category'] = 'Earthquake'
crisisnlp_earthquake_df.loc[crisisnlp_earthquake_df.label == '0', "Category"] = 'None'
crisisnlp_earthquake_df.rename(columns = {'text': 'tweet_texts', 'label': 'Disaster_or_not'}, inplace = True)
crisisnlp_earthquake_df = crisisnlp_earthquake_df.astype({'Disaster_or_not': 'int64'})
```

```python
crisisnlp_earthquake_df
```

```python
crisisnlp_earthquake_df.dtypes
```

## Flood


```python
crisisnlp_flood_df = pd.read_csv("CrisisNLP/floods_crisisnlp.tsv", sep='\t')
```

```python
crisisnlp_flood_df = crisisnlp_flood_df[crisisnlp_flood_df['lang'] == 'en']
crisisnlp_flood_df = crisisnlp_flood_df[['text', 'lable']]
crisisnlp_flood_df['lable'].replace({'non-eyewitness': '1', "don't know": '0', 'indirect-eyewitness': '1', 
                                          'direct-eyewitness': '1', 'vulnerable direct-eyewitness': '1'}, inplace = True)
crisisnlp_flood_df['Category'] = 'Flood'
crisisnlp_flood_df.loc[crisisnlp_flood_df.lable == '0', 'Category'] = 'None'
crisisnlp_flood_df.rename(columns = {'text': 'tweet_texts', 'lable': 'Disaster_or_not'}, inplace = True)
crisisnlp_flood_df = crisisnlp_flood_df.astype({'Disaster_or_not': 'int64'})
crisisnlp_flood_df
```

```python
crisisnlp_flood_df.dtypes
```

## Hurricanes

```python
crisisnlp_hurricane_df = pd.read_csv("CrisisNLP/hurricanes_crisisnlp.tsv", sep='\t')
```

```python
crisisnlp_hurricane_df = crisisnlp_hurricane_df[crisisnlp_hurricane_df['lang'] == 'en']
crisisnlp_hurricane_df = crisisnlp_hurricane_df[['text', 'Lable']]
crisisnlp_hurricane_df['Lable'].replace({'non-eyewitness': '1', "don't know": '0', 'indirect-eyewitness': '1', 
                                          'direct-eyewitness': '1', 'vulnerable-direct witness': '1'}, inplace = True)
crisisnlp_hurricane_df['Category'] = 'Hurricane'
crisisnlp_hurricane_df.loc[crisisnlp_hurricane_df.Lable == '0', 'Category'] = 'None'
crisisnlp_hurricane_df.rename(columns = {'text': 'tweet_texts', 'Lable': 'Disaster_or_not'}, inplace = True)
crisisnlp_hurricane_df = crisisnlp_hurricane_df.astype({'Disaster_or_not': 'int64'})
crisisnlp_hurricane_df
```

```python
crisisnlp_hurricane_df.dtypes
```

## Fire

```python
crisisnlp_fire_df = pd.read_csv("CrisisNLP/fires_crisisnlp.tsv", sep='\t')
```

```python
crisisnlp_fire_df = crisisnlp_fire_df[crisisnlp_fire_df['please_choose_a_category_:confidence'] > 0.6]
crisisnlp_fire_df = crisisnlp_fire_df[['text', 'please_choose_a_category_']]
crisisnlp_fire_df['please_choose_a_category_'].replace({'direct_eyewitness': '1', "dont_know": '0',
                                                        'noneyewitness': '1'}, inplace = True)
crisisnlp_fire_df['Category'] = 'Fire'
crisisnlp_fire_df.loc[crisisnlp_fire_df.please_choose_a_category_ == '0', 'Category'] = 'None'
crisisnlp_fire_df.rename(columns = {'text': 'tweet_texts', 'please_choose_a_category_': 'Disaster_or_not'}, inplace = True)
crisisnlp_fire_df = crisisnlp_fire_df.astype({'Disaster_or_not': 'int64'})
crisisnlp_fire_df
```

```python
crisisnlp_fire_df.dtypes
```

## Volcano

```python
crisisnlp_volcano_df = pd.read_csv("CrisisNLP/crisisnlp_volcano.csv")
```

```python
crisisnlp_volcano_df = crisisnlp_volcano_df[[' tweet_text', 'label']]
crisisnlp_volcano_df['label'].replace({'Not related to crisis': '0', 'Informative': '1', 
                                       'Not informative': '0', 'Non-government': '0'}, inplace = True)
crisisnlp_volcano_df['Category'] = 'Volcano'
crisisnlp_volcano_df.loc[crisisnlp_volcano_df.label == '0', 'Category'] = 'None'
crisisnlp_volcano_df.rename(columns = {' tweet_text': 'tweet_texts', 'label': 'Disaster_or_not'}, inplace = True)
crisisnlp_volcano_df = crisisnlp_volcano_df.astype({'Disaster_or_not': 'int64'})
crisisnlp_volcano_df.drop([3,4], axis = 0, inplace = True)
crisisnlp_volcano_df
```

```python
crisisnlp_volcano_df.dtypes
```

## Tornado

```python
crisisnlp_tornado_df = pd.read_csv("CrisisNLP/crisisnlp_tornado.csv")
```

```python
crisisnlp_tornado_df.drop(1047, axis = 0, inplace = True)
crisisnlp_tornado_df = crisisnlp_tornado_df[crisisnlp_tornado_df['choose_one:confidence'] > 0.5]
crisisnlp_tornado_df = crisisnlp_tornado_df[['text_no_rt', 'predicted']]
crisisnlp_tornado_df['predicted'].replace({'Informative (Indirect)': '1', 'Informative (Direct)': '1',
                                          'Personal message': '0', 'Other': '0'}, inplace = True)
crisisnlp_tornado_df['Category'] = 'Tornado'
crisisnlp_tornado_df.loc[crisisnlp_tornado_df.predicted == '0', 'Category'] = 'None'
crisisnlp_tornado_df.rename(columns = {'text_no_rt': 'tweet_texts', 'predicted': 'Disaster_or_not'}, inplace = True)
crisisnlp_tornado_df = crisisnlp_tornado_df.astype({'Disaster_or_not': 'int64'})
crisisnlp_tornado_df
```

```python
crisisnlp_tornado_df.dtypes
```

## Concatenate CrisisNLP datasets

```python
frames = [crisisnlp_earthquake_df, crisisnlp_flood_df, crisisnlp_fire_df, crisisnlp_hurricane_df, 
          crisisnlp_volcano_df, crisisnlp_tornado_df]
crisisnlp_disaster_df = pd.concat(frames)
crisisnlp_disaster_df = crisisnlp_disaster_df.replace('\n', '', regex = True)
crisisnlp_disaster_df = crisisnlp_disaster_df.sample(frac = 1)
crisisnlp_disaster_df
```
